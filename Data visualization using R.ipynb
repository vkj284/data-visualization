# Data visualization

### pie chart

```r
value <- c(45, 90, 30)
label <- c("mango", "orange", "apple")
```

```r
pie(value,label)
```

## bar chart

```r
H <- c(2,4,6,5,4)
barplot(H)
```

```r
H <- c(2,4,6,5,4)
M <- c("Mar","Apr","May","Jun","Jul")
```

```r
barplot(H,names.arg=M,xlab="Month",ylab="Revenue",col="blue",main="Revenue chart",border="red")
```

## Histograms

```r
H <- c(9,13,21,8,36,22,12,41,31,33,19)
```

```r
hist(H,xlab = "Weight",ylab = 'Frequencys',col = "blue",border = "yellow",main = 'histograms')
```

## Line - single Line

```r
a = c(7,14,21,24,20,15,10,5)
```

```r
plot(a,type = "o", col = "red", xlab = "Month", ylab = "Rain fall", main = "Rain fall chart")
```

## Line - Multiple line

```r
a = c(7,14,21,24,20,15,10,5)
b = c(5,6,9,7,7)
```

```r
plot(a,type = "o", col = "red", xlab = "Month", ylab = "Rain fall", main = "Rain fall chart")
lines(b, type = "o", col = "blue")
```

## Scatter Plot

```r
x = c(1,2,3,4,5)
y = c(5,4,3,2,1)
```

```r
plot(x, y, main ='Scatter Plot', xlab = 'x-axis', ylab = 'yaxis',col = 'blue',xlim = c(0,10),ylim = c(0,10) )
```
